DECISION TREE - REGRESSION MODEL

```python
#Importing required functions
import numpy as np
import pandas as pd
```

```python
#file upload
df=pd.read_csv("/content/Advertising.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
#check for missing values
df.isna().sum()
```

```python
#seperating x and y
x=df.iloc[:,:-1]
#x
y=df.iloc[:,-1]
#y
```

```python
#split data as x_train,x_test,y_train,y_test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
#model creation
from sklearn.tree import DecisionTreeRegressor
```

```python
#object creation
model=DecisionTreeRegressor()
```

```python
#model fit
model.fit(x_train,y_train)
```

```python
#pred
y_pred=model.predict(x_test)
y_pred
```

```python
#creation of dataframe using y_pred and y_test (predicted value and actual value)
df1=pd.DataFrame({"Actual Value":y_test,"Predicted Value":y_pred})
df1
```

```python
#Model evealuation - mean_absolute_error
from sklearn.metrics import mean_absolute_error
print("MAE = ",mean_absolute_error(y_test,y_pred))
```

```python
#Model evealuation - Mean Absolute Error %
from sklearn.metrics import mean_absolute_percentage_error
print("MAE % = ",mean_absolute_percentage_error(y_test,y_pred))
```

```python
#Model evealuation - R2 Score
from sklearn.metrics import r2_score
print("R2Score = ",r2_score(y_test,y_pred))
```

```python
#ploting decision tree
#import tree and plt

from sklearn import tree
import matplotlib.pyplot as plt
```

```python
#ploting decision tree
plt.figure(figsize=(15,15))
tree.plot_tree(model)
```
